 # Exploratory Data Analysis

```python
salla.columns
```

```python
genderwise=sb.countplot(x='Gender',data=salla)

for bars in genderwise.containers:
    genderwise.bar_label(bars)
```

*By the above graph we can see most of the buyers are Womens*

```python
salla.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=salla.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Gender',y='Amount',data=sales_gen)
```

*by the above graph we can understand that Womens are the major contributers of sales*

```python
genderwise = sb.countplot(data=salla,x='Age Group',hue='Gender')
for bars in genderwise.containers:
    genderwise.bar_label(bars)
```

```python
salla.columns
```

```python
sales_age=salla.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Age Group',y='Amount',data=sales_age)
```

*from the above graphs we can see most of the buyers are age group between 26-35 yrs female*

```python
sales_State=salla.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(x='State',y='Amount',data=sales_State)
```

```python
sales_orders=salla.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(17,5)})
sb.barplot(x='State',y='Orders',data=sales_orders)
```

 *by the above graphs we can see that uttar pradesh,maharastra and karnataka are the top performing states*

```python
genderwise=sb.countplot(x='Marital_Status',data=salla)    
sb.set(rc={'figure.figsize':(3,3)})
for bars in genderwise.containers:
    genderwise.bar_label(bars)
```

```python
sales_gen_mar=salla.groupby(['Marital_Status','Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(5,5)})
sb.barplot(x='Marital_Status',y='Orders',hue='Gender',data=sales_gen_mar)
```

```python
sales_mar_gen=salla.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(5,5)})
sb.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_mar_gen)
```

*from the above graphs we can see most of the buyer are married women and they have high purchasing power*

```python
salla.columns
```

```python
sb.set(rc={'figure.figsize':(20,5)})
genderwise=sb.countplot(x='Occupation',data=salla)
for bars in genderwise.containers:
    genderwise.bar_label(bars)
```

```python
salesstate=salla.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(17,5)})
sb.barplot(x ='Occupation',y='Amount',data=salesstate)
```

*from the above graph we can see mosgt of the buyers are from IT,healthcare and aviation*

```python
sb.set(rc={'figure.figsize':(20,5)})
genderwise=sb.countplot(x='Product_Category',data=salla)
for bars in genderwise.containers:
    genderwise.bar_label(bars)
```

```python
salesbyproduct=salla.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(x='Product_Category',y='Amount',data=salesbyproduct)
```

*from the above graph we can see most of the sold products are Food,Clothingg&apparel and Electronics*

```python
salesbyorders=salla.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(x='Product_ID',y='Orders',data=salesbyorders)
```

*from above graph we can see that product id of poo265242,p00110942 and p00237542 are the top selling products*
